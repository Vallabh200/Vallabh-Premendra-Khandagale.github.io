```python
#Import desired libararies.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
#Read Data set
df = pd.read_csv('/content/drive/MyDrive/NSM PRACTICAL/MINI PROJECT/Group C/tool_wear_dataset - tool_wear_dataset.csv')
```

## **Model preparation - Simple Linear Regression**
________________________________________________________________________________  __________


```python
#Coloumn Name
df.columns
```

```python
# Define X and y
X = df[['time']].values
y = df['tool_wear'].values
```

```python
# Define model and fit for X,y.
lin_model = LinearRegression()
lin_model.fit(X, y)
```

```python
# Equation of linear model, tool wear =  *time+b
a=lin_model.coef_
b=lin_model.intercept_
print("\nEquation of Linear model\n")
print(f"Tool_wear = ({a[0]:.4f}) * time + ({b:.4f})\n")

#make the underlione
print("*"*50)
```

```python
# Prediction of tool wear
y_pred = lin_model.predict(X)
```

```python
# Plot a linear Regression curve to data set
plt.figure(figsize=(7,5))
plt.scatter(df["time"], df["tool_wear"], color="blue",s=10,label='Data')
plt.plot(df["time"], y_pred, color="red", label='Linear Regression')
plt.xlabel("Time")
plt.ylabel("Tool Wear")
plt.title("Linear Regression Model for Tool Wear Prediction")
plt.grid(True)
plt.legend()
plt.show()
```

#Evaluate Linear regression model - r^2 , RMSE.

```python
# Evaluate Linear regression model - r^2 and RMSE.
r2_lin = r2_score(y, y_pred)
rmse_lin = np.sqrt(mean_squared_error(y, y_pred))

print(f"R-squared (Linear Regression): {r2_lin:.4f}")
print(f"RMSE (Linear Regression): {rmse_lin:.4f}")

# Plotting residuals for linear regression
residuals_lin = y - y_pred
plt.figure(figsize=(7, 5))
plt.scatter(y_pred, residuals_lin, color="blue", s=10, alpha=0.5)
plt.xlabel("Predicted Tool Wear")
plt.ylabel("Residuals")
plt.title("Residuals Plot for Linear Regression Model")
plt.axhline(y=0, color='red', linestyle='--')
plt.grid(True)
plt.show()
```

## ***Multiple Regression***

```python
# Define Xm and y
Xm = df[['cutting_speed', 'feed_rate', 'depth_of_cut', 'time']].values
y = df['tool_wear'].values

# Define and fit the multiple regression model
multi_model = LinearRegression()
multi_model.fit(Xm, y)

# Print the equation of the multiple regression model
a_multi = multi_model.coef_
b_multi = multi_model.intercept_
print("\nEquation of Multiple Regression model\n")
print(f"Tool_wear = ({a_multi[0]:.4f}) * cutting_speed + ({a_multi[1]:.4f}) * feed_rate + ({a_multi[2]:.4f}) * depth_of_cut + ({a_multi[3]:.4f}) * time + ({b_multi:.4f})\n")
print("*"*50)

# Prediction of tool wear
y_pred_multi = multi_model.predict(Xm)

# Evaluate the multiple regression model - r^2 and RMSE.
r2_multi = r2_score(y, y_pred_multi)
rmse_multi = np.sqrt(mean_squared_error(y, y_pred_multi))

print(f"R-squared (Multiple Regression): {r2_multi:.4f}")
print(f"RMSE (Multiple Regression): {rmse_multi:.4f}")

# Plotting residuals for multiple regression
residuals_multi = y - y_pred_multi
plt.figure(figsize=(7, 5))
plt.scatter(y_pred_multi, residuals_multi, color="blue", s=10, alpha=0.5)
plt.xlabel("Predicted Tool Wear")
plt.ylabel("Residuals")
plt.title("Residuals Plot for Multiple Regression Model")
plt.axhline(y=0, color='red', linestyle='--')
plt.grid(True)
plt.show()
```
